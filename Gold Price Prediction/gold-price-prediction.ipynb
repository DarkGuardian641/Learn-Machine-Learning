### Splitting the Features and Target

```python
X = df.drop(["Date","GLD"], axis = 1)
Y = df["GLD"]
```

```python
# Print X
X
```

```python
# Print Y
Y
```

### Splitting the dataset into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 3)
```

### Model Training : Random Forest Regressor

```python
regressor = RandomForestRegressor(n_estimators = 100)
```

```python
# Training the model
regressor.fit(X_train, Y_train)
```

### Model Evaluation

```python
# Prediction on test data
test_data_prediction = regressor.predict(X_test)
```

```python
# Print test_data_prediction
test_data_prediction
```

```python
# R squared error
error_score = metrics.r2_score(Y_test,test_data_prediction)
print("R squared error : ", error_score)
```

### Compare actual values and predicted values

```python
Y_test = list(Y_test)
```

```python
plt.figure(figsize=(12, 8))  # Added this line to make plot bigger
plt.plot(Y_test, color = "blue", label = "Actual value")
plt.plot(test_data_prediction, color = "green", label = "Predicted value")
plt.title("Actual price vs Predicted price")
plt.xlabel("Number of values")
plt.ylabel("Gold price")
plt.legend(loc='upper right')
plt.show()
```

```python
# Function to predict gold price based on input values
def predict_gold_price():
    print("Enter the following market indicators to predict gold price:")
    spx = float(input("Enter SPX (Standard & Poor's 500) value: "))
    uso = float(input("Enter USO (United States Oil Fund) value: "))
    slv = float(input("Enter SLV (Silver) value: "))
    eur_usd = float(input("Enter EUR/USD exchange rate: "))
    
    # Print entered values
    print("\nEntered Values:")
    print(f"SPX: {spx}")
    print(f"USO: {uso}")
    print(f"SLV: {slv}")
    print(f"EUR/USD: {eur_usd}")
    
    # Create input data with proper feature names
    input_data = pd.DataFrame({
        'SPX': [spx],
        'USO': [uso],
        'SLV': [slv],
        'EUR/USD': [eur_usd]
    })
    
    # Make prediction
    prediction = regressor.predict(input_data)
    
    print(f"\nPredicted Gold Price: ${prediction[0]:.2f}")

# Example usage
print("Gold Price Prediction System")
print("---------------------------")
predict_gold_price()
```
