1. Implement logistic regression using Python/R to perform classification on Social_Network_Ads.csv dataset. 
2. Compute Confusion matrix to find TP, FP, TN, FN, Accuracy, Error rate, Precision, Recall on the given dataset.

```python
# Import the dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
# Load the dataset
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
# Print first 5 rows
df.head()
```

```python
# Columns
df.columns
```

```python
# Quantitative to Categorical data
df['Gender'].replace({"Male":0, "Female":1},inplace=True)
```

```python
# Print dataframe again to check Gender Column
df.head()
```

```python
# Split the dataframe
# You can also use this : x = df.drop(['Purchased'], axis=1)
x= df[['User ID','Gender','Age','EstimatedSalary']]
y=df[['Purchased']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25,  random_state = 29)
```

```python
# Logistic Regression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# Prediction
y_predict = model.predict(x_test)
y_predict
```

```python
# Confusion Matrix
cm=confusion_matrix(y_test, y_predict)
cm
```

```python
# Print condusion matrix docs
print(confusion_matrix.__doc__)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_predict).ravel()
```

```python
# Accuracy Score
a = accuracy_score(y_test, y_predict)
print("accuracy score is: ", a)
```

```python
# Error
e= 1-a
print("error is: ", e)
```

```python
# Precision
p= precision_score(y_test, y_predict)
print("precision score is: ", p)
```

```python
# Recall Score
r= recall_score(y_test, y_predict)
print("recall score is", r)
```
